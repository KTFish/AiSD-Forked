# Lab 9 - sortowanie

Metody sortowania są jednymi z najpopularniejszych algorytmów w informatyce.
Ich celem jest uporządkowanie pewnego zbioru danych względem pewnego charakterystycznego atrybutu.
Zbiór można również posortować względem wartości każdego elementu, jeżeli występuje taka możliwość.
Przykładem takiego zbioru może być zbiór liczb.
Głównym celem sortowania zbiorów jest ich uporządkowanie celem przekazania
innym algorytmom (np. wyszukującym dane), lub prezentacja w formie czytelnej dla człowieka.

## Sortowanie bąbelkowe (bubble sort)

Złożoność obliczeniowa: $O(N^2)$

Metoda sortowania bąbelkowego polega na porównywaniu dwóch elementów i zamianie ich kolejności,
jeżeli zaburza porządek docelowy.

### Przykład (sortowanie listy liczb całkowitych)

Wejście: [6, 3, 9, 0, 4]

Cel: sortowanie tablicy wejściowej w porządku rosnącym

Dokonujemy n-1 kroków "wybąblowywania" elementów w przód 
w tablicy, gdzie n oznacza liczbę elementów w tej tablicy.
Po każdym każdym pojedynczym "wybąblowaniu" zmniejszamy
rozpatrywany przedział o 1 element, aż dojdziemy do tablicy
o długości 1.

Krok 1: bąblujemy tablicę od el. o indeksie 0 po n-1 w przód
- lista[0]=6, lista[1]=3, 6 > 3 -> [3, 6, 9, 0, 4]
- lista[1]=6, lista[2]=9, 6 <=9 -> [3, 6, 9, 0, 4] (bez zmian)
- lista[2]=9, lista[3]=0, 9 > 0 -> [3, 6, 0, 9, 4]
- lista[3]=9, lista[4]=9, 9 > 4 -> [3, 6, 0, 4, 9]

Krok 2: bąblujemy tablicę od el. o indeksie 0 po n-2 w przód
- lista[0]=3, lista[1]=6, 3 <=6 -> [3, 6, 0, 4, 9] (bez zmian)
- lista[1]=6, lista[2]=0, 6 > 0 -> [3, 0, 6, 4, 9]
- lista[2]=6, lista[3]=4, 6 > 4 -> [3, 0, 4, 6, 9]

Krok 3:  bąblujemy tablicę od el. o indeksie 0 po n-3 w przód
- lista[0]=3, lista[1]=0, 3 > 0 -> [0, 3, 4, 6, 9]
- lista[1]=3, lista[2]=4, 3 <=4 -> [0, 3, 4, 6, 9] (bez zmian)

Krok 4: bąblujemy tablicę od el. o indeksie 0 po n-4 w przód
- lista[0]=0, lista[1]=3, 0 <=3 -> [0, 3, 4, 6, 9] (bez zmian)  

Wyjscie: [0, 3, 4, 6, 9]

### Pseudokod:

- niech tab będzie listą/tablicą wejściową (nieposortowaną)
- niech n będzie liczbą elementów w tab
- pętla for po wszystkich indeksach tab, gdzie zmienną iterującą będzie k zaczynająca od n-1 i zmniejszajaca się do 1 (włącznie)
    - pętla for po wszystkich indeksach l, od 0 po k-1 (włącznie):
        - jeżeli tab[l] > tab[l+1], to zamień te wartości miejscami

## Sortowanie przez wybieranie (selection sort)

Złożoność obliczeniowa: $O(N^2)$

Metoda sortowania przez wybieranie polega na wybieraniu najmniejszych wartości i wstawianiu ich na początku tabeli.
Po każdym kroku ta procedura zaczyna się od kolejnego elementu, aż dojdzie się do 1-elementowej tabeli.

### Przykład (sortowanie listy liczb całkowitych)

Wejście: tab = [6, 3, 9, 0, 4]

Cel: sortowanie tablicy wejściowej w porządku rosnącym

Dokonujemy n-1 kroków wyszukiwania elementów najmniejszych w tabeli o indeksach k..n-1.
Następnie element ten zamieniamy z elementem tab[k]. Po każdym kroku k zwiększamy o 1.

Krok 0: rozpatrywana tab[0..n-1]
- imin = 3 = indeks najmniejszego el. w tab[0..n-1] 
- zamiana elementów tab[0] z tab[imin] -> [0, 3, 9, 6, 4]

Krok 1: rozpatrywana tab[1..n-1]
- imin = 1 = indeks najmniejszego el. w tab[1..n-1] 
- imin == 1 -> brak zamiany 

Krok 2: rozpatrywana tab[2..n-1]
- imin = 4 = indeks najmniejszego el. w tab[2..n-1] 
- zamiana elementów tab[2] z tab[imin] -> [0, 3, 4, 6, 9]

Krok 3: rozpatrywana tab[3..n-1]
- imin = 3 = indeks najmniejszego el. w tab[3..n-1] 
- imin == 3 -> brak zamiany 

### Pseudokod:
- niech tab będzie tablicą wejściową (nieposortowaną)
- niech n będzie liczbą elementów w tab
- pętla for po indeksach k = od 0 po n-2(włącznie):
    - imin = indeks najmniejszego elementu w tab[k..n-1]
    - jeśli imin != k to zamień tab[k] z tab[imin]

## Sortowanie przez wstawianie (insertion sort) z zaawansowanym przesuwaniem

Złożoność obliczeniowa: $O(N^2)$

Metoda sortowania przez wstawianie polega na wstawianiu do stale rosnącej
tablicy z posortowanymi wartościami (tab[0..k-1]) nowego elementu (tab[k]) 
w taki sposób, aby elementy były wciąż posortowane (tab[0..k]).

### Przykład (sortowanie tablicy liczb całkowitych)

Wejście: tab = [6, 3, 9, 0, 4]

Cel: sortowanie listy wejściowej w porządku rosnącym

Krok 1:  dodanie do posortowanej tablicy tab[0..0] elementu tuż za tą pod-tablicą, czyli tab[1]
- niech k = 1
- key = tab[k] = tab[1] = 3
- niech m = 0 = indeks pierwszego elementu w tab[0..k-1] o wartości większej niż key
 ; gdyby wszystkie elementy w tab[0..k-1] <= key, wtedy m = k 
- m != k, więc pracuj dalej
- przesuń w prawo tab[m..k-1] -> [6, 6, 9, 0, 4]
- tab[m] = key = 3 -> [3, 6, 9, 0, 4]

Krok 2: dodanie do posortowanej tablicy tab[0..1] elementu tuż za tą pod-tablicą, czyli tab[2]
- niech k = 2
- key = tab[k] = tab[2] = 9
- niech m = 2 = indeks pierwszego elementu w tab[0..k-1] o wartości większej niż key
 ; gdyby wszystkie elementy w tab[0..k-1] <= key, wtedy m = k 
- m == k, więc nie rób czegokolwiek więcej w tym kroku

Krok 3: dodanie do posortowanej tablicy tab[0..2] elementu tuż za tą pod-tablicą, czyli tab[3]
- niech k = 3
- key = tab[k] = tab[3] = 0
- niech m = 0 = indeks pierwszego elementu w tab[0..k-1] o wartości większej niż key
 ; gdyby wszystkie elementy w tab[0..k-1] <= key, wtedy m = k 
- m != k, więc pracuj dalej
- przesuń w prawo tab[m..k-1] -> [3, 3, 6, 9, 4]
- tab[m] = key = 0 -> [0, 3, 6, 9, 4]

Krok 4: dodanie do posortowanej tablicy tab[0..3] elementu tuż za tą pod-tablicą, czyli tab[4]
- niech k = 4
- key = tab[k] = tab[4] = 4
- niech m = 2 = indeks pierwszego elementu w tab[0..k-1] o wartości większej niż key
 ; gdyby wszystkie elementy w tab[0..k-1] <= key, wtedy m = k 
- m != k, więc pracuj dalej
- przesuń w prawo tab[m..k-1] -> [0, 3, 6, 6, 9]
- tab[m] = key = 4 -> [0, 3, 4, 6, 9]

### Pseudokod
- niech tab będzie tablicą/listą wejściową (nieposortowaną)
- niech n będzie liczbą elementów w tab
- pętla for ze zmienną iterującą k w zakresie 1..n-1(włącznie)
    - key = tab[k]
    - m = indeks pierszego elementu w tab[0..k-1] > key, bądź m = k jeśli wszystkie el. w tab[0..k-1] są mniejsze lub równe
    - jeśli m == k, wtedy przeskocz resztę czynności w pętli (continue)
    - przesuń w prawo wszystkie elementy w tab[m..k-1], czyli tab[m+1..k] = tab[m..k-1];
        uwaga, tutaj łatwo o stworzenie błędu
    - tab[m] = key
