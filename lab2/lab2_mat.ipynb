# Lab 2 - sortowanie

Metody sortowania są jednymi z najpopularniejszych algorytmów w informatyce.
Ich celem jest uporządkowanie pewnego zbioru danych względem pewnego charakterystycznego atrybutu.
Zbiór można również posortować względem wartości każdego elementu, jeżeli występuje taka możliwość.
Przykładem takiego zbioru może być zbiór liczb.
Głównym celem sortowania zbiorów jest ich uporządkowanie celem przekazania
innym algorytmom (np. wyszukującym dane), lub prezentacja w formie czytelnej dla człowieka.

## Sortowanie bąbelkowe (bubble sort)

Metoda sortowania bąbelkowego polega na porównywaniu dwóch elementów i zamianie ich kolejności,
jeżeli zaburza porządek docelowy.

### Przykład (sortowanie listy liczb całkowitych)

Wejście: [6, 3, 9, 0, 4]

Cel: sortowanie listy wejściowej w porządku rosnącym

Dokonujemy n kroków porównania każdej pary liczb
w liście, gdzie n oznacza liczbę elementów w tej liście.

Krok 1: porównanie każdej pary liczb w liście
- (6, 3): 6 > 3 -> [3, 6, 9, 0, 4]
- (3, 9): 3 < 9 -> [3, 6, 9, 0, 4] (lista bez zmian)
- (3, 0): 3 > 0 -> [0, 6, 9, 3, 4]
- (0, 4): 0 < 4 -> [0, 6, 9, 3, 4] (lista bez zmian)

Krok 2: kolejne porównanie par liczb w liście
- (6, 9): 6 < 9 -> [0, 6, 9, 3, 4] (lista bez zmian)
- (6, 3): 6 > 3 -> [0, 3, 9, 6, 4]
- (3, 4): 3 < 4 -> [0, 3, 9, 6, 4] (lista bez zmian)

Krok 3: kolejne porównanie par liczb w liście
- (9, 6): 9 > 6 -> [0, 3, 6, 9, 4]
- (6, 4): 6 > 4 -> [0, 3, 4, 9, 6]

Krok 4: kolejne porównanie par liczb w liście
- (9, 6): 9 > 6 -> [0, 3, 4, 6, 9]

Wyjscie: [0, 3, 4, 6, 9]

### Pseudokod:
Sortowanie rosnące metodą bąbelkową

- niech l będzie listą wejściową (nieposortowaną)
- pętla for po wszystkich indeksach l, gdzie zmienną iterującą będzie i:
    - pętla for po wszystkich indeksach l, gdzie zmienną iterującą będzie j:
        - jeżeli l[i] < l[j], to zamień te wartości miejscami w liście

## Sortowanie przez wybieranie (selection sort)

Metoda sortowania przez wybieranie polega na wskazywaniu kolejnych pozycji na liście
wejściowej, a następnie na wyszukiwaniu najmniejszego elementu w liście z wyjątkiem wskazanej
pozycji i zamianie tych wartości, jeżeli ich ułożenie zaburza porządek.

### Przykład (sortowanie listy liczb całkowitych)

Wejście: l = [6, 3, 9, 0, 4]

Cel: sortowanie listy wejściowej w porządku rosnącym

Dokonujemy n kroków porównania każdej pary liczb
w liście, gdzie n oznacza liczbę elementów w tej liście.
W każdym z tych porównań dla pozycji i=0, 1, 2, ..., n - 1
wstawiamy najmniejszą znalezioną wartość w pozycjach od i + 1.

Krok 0: porównania par liczb w liście
- index = 0
- (l[1], l[index]): 3 < 6 -> index = 1
- (l[2], l[index]): 9 > 3 -> index = 1 (bez zmian)
- (l[3], l[index]): 0 < 3 -> index = 3
- (l[4], l[index]): 4 > 0 -> index = 3 (bez zmian)
- l = [0, 3, 9, 6, 4] (dokonujemy zamiany wartości na pozycji równej numerowi kroku i pozycji index i w ten sposób na początku nowej listy znajduje się najmniejsza wartość)

Krok 1: kolejne porównania par liczb
- index = 1
- (l[2], l[index]): 9 > 3 -> index = 1 (bez zmian)
- (l[3], l[index]): 6 > 3 -> index = 1 (bez zmian)
- (l[4], l[index]): 4 > 3 -> index = 1 (bez zmian)
- l = [0, 3, 9, 6, 4] (lista pozostaje bez zmian - zamiana liczb na pozycji 1)

Krok 2: kolejne porównania par liczb
- index = 2
- (l[3], l[index]): 6 < 9 -> index = 3
- (l[4], l[index]): 4 < 6 -> index = 4
- l = [0, 3, 4, 6, 9] (lista jest już posortowana)

Krok 3: kolejne porównania par liczb
- index = 3
- (l[4], l[index]): 9 > 6 -> index = 3 (bez zmian)
- l = [4, 0, 3, 6, 9] (lista pozostaje bez zmian, finalne ułożenie nastąpiło w kroku 3)

### Pseudokod:
- niech l będzie listą wejściową (nieposortowaną)
- pętla for po wszystkich indeksach l, gdzie zmienną iterującą będzie i:
    - min_index = i
    - pętla for w zakresie od i + 1 do długości l - 1:
        - jeżeli l[j] < l[min_index] to ustaw min_index = j
    - zamień miejscami wartości w liście na pozycjach j oraz min_index

## Sortowanie przez wstawianie (insertion sort)

Metoda sortowania przez wstawianie polega na ustawianiu elementów w odpowiednich miejscach, jednocześnie przesuwając pozostałe.

### Przykład (sortowanie listy liczb całkowitych)

Wejście: l = [6, 3, 9, 0, 4]

Cel: sortowanie listy wejściowej w porządku rosnącym

W każdym kroku dokonujemy operacji modyfikacji listy elementów dopóki j >= 0 oraz l[j] > key

Krok 1:
- key = l[1] (3)
- j = 0
- l = [6, 6, 9, 0, 4] (na pozycji j+1 w l wstawiamy wartośc z l na pozycji j)
- j = j - 1 (-1)
- Krok dobiega końca, gdyż j < 0 lub l[j] <= key
- l[j+1] = key (l = [3, 6, 9, 0, 4])

Krok 2:
- key = l[2] (9)
- j = 1
- krok dobiega końca gdyż 6 < 9 (l[j] <= key)
- l[j+1] = key (l = [3, 6, 9, 0, 4])

Krok 3:
- key = l[3] (0)
- j = 2
- l = [3, 6, 9, 9, 4]
- j = j - 1 (1)
- l = [3, 6, 6, 9, 4]
- j = j - 1 (0)
- l = [3, 3, 6, 9, 4]
- j = j - 1 (-1)
- Krok dobiega końca, gdyż j < 0
- l[j+1] = key (l = [0, 3, 6, 9, 4])

Krok 4:
- key = l[4] (4)
- j = 3
- l = [0, 3, 6, 9, 9]
- j = j - 1 (2)
- l = [0, 3, 6, 6, 9]
- j = j - 1 (1)
- Krok dobiega końca, gdyż 3 < 4 (l[j] <= key)
- l[j+1] = key (l = [0, 3, 4, 6, 9])

### Pseudokod
- niech l będzie listą wejściową (nieposortowaną)
- niech n będzie liczbą elementów w l
- pętla for ze zmienną iterującą i w zakresie od 1 do n - 1:
    - key = l[i]
    - j = i - 1
    - while j >= 0 and l[j] > key:
        - l[j + 1] = l[j]
        - j = j - 1
    l[j + 1] = key
